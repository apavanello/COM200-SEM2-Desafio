```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
URI = "https://raw.githubusercontent.com/fivethirtyeight/data/master/classic-rock/classic-rock-song-list.csv"
ds = pd.read_csv(URI, on_bad_lines='skip' ,encoding='utf-8')

#remove as musicas sem ano
ds = ds.dropna( axis=0, how='any' )

#limpa o dataset
ds = ds.drop(columns=['COMBINED', 'First?','Year?','F*G'])

#traduz os nomes das colunas
ds.columns = ['Musica', 'Artista', 'Ano de Lancamento', 'qnt de vezes tocada']

#remove valores nao inteiros da coluna Ano de Lancamento
ds = ds[ds['Ano de Lancamento'].apply(lambda x: str(x).isdigit())]

#remove anos menores que 1950 e maiores que 2024
ds = ds[ds['Ano de Lancamento'].apply(lambda x: int(x) > 1950 and int(x) < 2024)]
```

```python
#Pega o indice da musica mais tocada por ano
indice_da_musica_mais_tocada_por_ano = ds.groupby('Ano de Lancamento')['qnt de vezes tocada'].idxmax()

#Localiza o indice no dataset e volta a linha 
musicas_mais_tocadas_por_ano = ds.loc[indice_da_musica_mais_tocada_por_ano]

#Classifica o dataset por ano em ordem crescente
musicas_mais_tocadas_por_ano_organizada = musicas_mais_tocadas_por_ano.sort_values(by='Ano de Lancamento')

```

```python
#Mostra o ds organizado por ano com a musica mais tocada com as colunas Ano de lancamento, musica e artista e esconde o indice

musicas_mais_tocadas_por_ano_organizada[['Ano de Lancamento', 'Musica', 'Artista']].style.hide()
```
